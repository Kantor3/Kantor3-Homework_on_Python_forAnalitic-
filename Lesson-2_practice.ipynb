- sqft_living15 - Квадратные метры внутренней жилой площади жилья для ближайших 15 соседей

## 4 Задача
Ответьте на несколько вопросов:
1) Сколько в среднем стоят дома, у которых 2 спальни?
2) Какая в среднем общая площадь домов, у которых стоимость больше 600 000?
3) Как много домов коснулся ремонт?
4) Насколько в среднем стоимость домов с оценкой grade домов выше 10 отличается от стоимости домов с оценкой grade меньше 4?

### 4.1 Сколько в среднем стоят дома, у которых 2 спальни?

```python
tmp_bedrooms = 2
mean_price_bedrooms = round(df[df.bedrooms == tmp_bedrooms].price.mean())
print(f'Дома, у которых {tmp_bedrooms} спальни, в среднем стоят {mean_price_bedrooms} USD')
```

### 4.2 Какая в среднем общая площадь домов, у которых стоимость больше 600 000?

```python
tmp_price = 600000
mean_sqft_price = round(df[df.price > tmp_price].sqft_lot.mean())
print(f'Площадь домов, у которых стоимость больше {tmp_price} USD в среднем имеют общую площать = {mean_sqft_price} м2')
```

### 4.3 Как много домов коснулся ремонт?

```python
count_renovated, _ = df[~(df.yr_renovated == 0)].shape
print(f'Ремонт коснулся {count_renovated} домов')
```

### 4.4 Насколько в среднем стоимость домов с оценкой grade домов выше 10 отличается от стоимости домов с оценкой grade меньше 4?

```python
grade1 = 10
cost_grade1 = df[df.grade > grade1].price.mean()
grade2 = 4
cost_grade2 = df[df.grade < grade2].price.mean()
k_price = round(cost_grade1 / cost_grade2, 2) if cost_grade2 else 1
on_price = round(cost_grade1 - cost_grade2)
print(f'Стоимости домов с оценкой выше {grade1} и оценкой меньше {grade2} отличаюся:\n- в {k_price} раза \n- на {on_price} USD')
```

### 5.1 Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. Сколько вариантов есть у клиента?

```python
variants = set(df.waterfront.unique())
print(f'Варианты оценки вида на набережную: {variants}')
```

```python
txt_cond = 'C видом на набережную, как минимум с тремя ванными и с подвалом'
mask_51 = df.waterfront.astype('bool') & (df.bathrooms > 2) & df.sqft_basement.astype('bool')
variants, _ = df[mask_51].shape
print(f'Число вариантов домов [{txt_cond}] = {variants}')
print('\nВот эти варианты:')
df[mask_51]
```

### 5.2 Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?

```python
txt_cond = f'С очень красивым видом из окна, либо с видом на набережную,' \
           f'в очень хорошем состоянии и год постройки не меньше 1980 года'

view_max = df.view.max()
state_max = df.condition.max()

df_cond = df[(df.view == view_max) | 
             (df.waterfront == 1) & (df.condition == state_max) & (df.yr_built >= 1980)]

variants, _ = df_cond.shape
cost_min = df_cond.price.min()
cost_max = df_cond.price.max()

print(f'Дома [{txt_cond}] \n идут по цене от {cost_min} до {cost_max} USD и таких вариантов = {variants}')
print('\nВот эти варианты:')
df_cond
```

### 5.3 Клиент хочет дом без подвала, с двумя этажами, стоимостью до 150000. Какая оценка по состоянию у таких домов в среднем?

```python
txt_cond = 'Дома без подвала, с двумя этажами, стоимостью до 150000'
mask_cond = (df.sqft_basement == 0) & (df.floors == 2) & (df.price < 150000)
df_cond = df[mask_cond]
variants, _ = df_cond.shape
print(f'[{txt_cond}] в среднем оцениваются на {df_cond.condition.mean()} баллов. \n Таких вариантов = {variants}')
print('\nВот эти варианты:')
df_cond
```
